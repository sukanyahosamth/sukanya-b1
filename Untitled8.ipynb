```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

df = sns.load_dataset('titanic')
df.describe()
```

```python
sns.countplot(x='pclass', data=df)
plt.title("Passenger Class Distribution")
plt.xlabel("(1 = Upper, 2 = Middle, 3 = Lower)")
plt.show()
```
